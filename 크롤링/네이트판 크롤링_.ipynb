```python
import requests
from scrapy.http import TextResponse
import csv
import re
import requests
import time
import pandas as pd
```

#1. 로스트아크 인벤

```python
# 페이지당 100게시물, (100개 * 50페이지 = 5000)
save=[]
for a in range(53, 2, -1): # 53페이지부터 3페이지까지 역순으로 반복
        url = 'https://www.inven.co.kr/board/lostark/4811?p='+str(a)
        req = requests.get(url)
        response = TextResponse(req.url, body=req.text, encoding='utf-8')
        for b in range (1,51):  # 각 페이지별 게시물 1번부터 50번까지 반복
                content = response.xpath(
                  '//*[@id="new-board"]/form/div/table/tbody/tr['+str(b)+']/td[2]/div/div/a/text()'
                 
                ).extract()[1].strip()
                save.append([content])
                print(content)
        time.sleep(15)
```

```python
df = pd.DataFrame(save,columns=['제목'])
df.to_csv('inven_lostark.csv',index=False,encoding='utf-8-sig')
```

#4. 네이트판 10대이야기

```python
# 페이지당 30게시물, 30개 * 330페이지 = 10000개
save=[]
for a in range(333, 2, -1):  # 333페이지부터 3페이지까지 역순으로 반복
        url = 'https://pann.nate.com/talk/c20038?page='+str(a)
        req = requests.get(url)
        response = TextResponse(req.url, body=req.text, encoding='utf-8')
        for b in range (1,31):   # 각 페이지별 게시물 1번부터 30번까지 반복
                content = response.xpath(
                  '//*[@id="searchDiv"]/div[1]/table/tbody/tr['+str(b)+']/td[1]/a/text()'  
                ).extract()[0].strip()
                save.append([content])
                print(content)
        time.sleep(10)
```

```python
df = pd.DataFrame(save,columns=['제목'])
df.to_excel('natepanteen.xls',index=False,encoding='utf-8-sig')
```
