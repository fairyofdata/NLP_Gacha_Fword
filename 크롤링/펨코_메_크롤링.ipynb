```python
import requests
from scrapy.http import TextResponse
import csv
import re
import requests
import time
import pandas as pd
import numpy as np
```

```python
save=[]

for a in range(252, 1, -1):
    r=np.random.randint(59,66)
    r2=np.random.randint(992,1100)
    r3=np.random.randint(5,6)
    rf=r*r2*r3*0.00009851204
    url_maple = 'https://www.fmkorea.com/index.php?mid=maple&page='+str(a)
    req_maple = requests.get(url_maple)  
    time.sleep(rf)   
    response_maple = TextResponse(req_maple.url, body=req_maple.text, encoding='utf-8')
    for b in range(1,20):  
        content_maple = response_maple.xpath(
            '//*[@id="bd_1135415169_0"]/div/table/tbody/tr['+str(b)+']/td[2]/a/text()'
                ).extract()[0].strip()
        save.extend([content_maple])
    print(a) 
```

```python
df = pd.DataFrame(save, columns=['제목'])
df.to_csv('fmmaple.csv', index=False, encoding='utf-8-sig') 
```
