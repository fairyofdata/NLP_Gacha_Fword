```python
save=[]

for a in range(252, 1, -1):
    r=np.random.randint(59,66)
    r2=np.random.randint(992,1100)
    r3=np.random.randint(10,12)
    rf=r*r2*r3*0.00009851204
    url_lol = 'https://www.fmkorea.com/index.php?mid=lol&page='+str(a)
    req_lol = requests.get(url_lol)
    time.sleep(rf)  
    response_lol = TextResponse(req_lol.url, body=req_lol.text, encoding='utf-8')
    for b in range(1,20):  
        content_lol = response_lol.xpath(
            '//*[@id="bd_14339012_0"]/div/table/tbody/tr['+str(b)+']/td[2]/a/text()'
                ).extract()[0].strip()
        save.extend([content_lol])
    print(a) 
```

```python
df = pd.DataFrame(save, columns=['제목'])
df.to_csv('fmlol.csv', index=False, encoding='utf-8-sig') 
```
