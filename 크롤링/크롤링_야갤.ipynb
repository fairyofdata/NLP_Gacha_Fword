```python
data=[]
for page in range(2, 503):
    url0='https://gall.dcinside.com/board/lists/?id=baseball_new11&page='
    url1=url0+str(page)
    response = requests.get(url1, headers=headers[0])
    soup = BeautifulSoup(response.content, 'html.parser')
    html_list = soup.find('tbody').find_all('tr')
    
    for i in html_list:
        title_list = i.find('a', href=True).text # 제목
        data.append([title_list])

len(data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```

```python
df['1']=0
```
