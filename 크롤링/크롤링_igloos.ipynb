```python
import requests
from scrapy.http import TextResponse
import csv
import re
import requests
import time
import pandas as pd
```

```python
save=[]
for a in range(88738,88777,1):
  url_egloo = 'http://pjj321.egloos.com/m/40'+str(a)
  req_egloo = requests.get(url_egloo)
  response_egloo = TextResponse(url_egloo, body=req_egloo.text, encoding='utf-8')
  for b in range(1,34):  
      content_egloo = response_egloo.xpath(
          '//*[@id="container"]/article/div['+str(b)+']/strong/text()'
          ).extract()
      save.extend([content_egloo])
print(save) 
```

```python
df = pd.DataFrame(save)
df.to_csv('egloo_test.csv', index=False, encoding='utf-8-sig') 
```
