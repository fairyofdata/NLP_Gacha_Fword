```python
url0='https://hygall.com/index.php?mid=hy&page=2'
```

```python
data=[]
for page in range(2, 10):
    url0='https://hygall.com/index.php?mid=hy&page='
    url1=url0+str(page)
    response = requests.get(url1, headers=headers[0])
    soup = BeautifulSoup(response.content, 'html.parser')
    html_list = soup.find('tbody').find_all('tr')
    
    for i in html_list:
        title_list = i.find('a', href=True).text # 제목
        data.append([title_list])

len(data)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data)
```

```python
df['1']=0
```

```python
df.columns=['records', 'label']
df.head(30)
```
